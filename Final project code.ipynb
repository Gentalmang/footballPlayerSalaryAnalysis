```python
import pandas as pd
import math as m
import re
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
```

```python
dataSet = pd.read_csv("Soccer_Salaries_EPL.csv")
```

```python
ages = []
sum_salaries = []
average_salaries = []

for i in range(22):
    sum_salaries.append(0)
    average_salaries.append(0)
    
for (age, salary) in zip(dataSet['Age'],dataSet['Weekly_Salary']):
    ages.append(age)

age_distribution = []
x = range(18,40)

for i in x:
    age_distribution.append(ages.count(i))
    
for (age, salary) in zip(dataSet['Age'],dataSet['Weekly_Salary']):
    for i in x:
        if age == i:
            sum_salaries[i-18] += int(re.sub("[^0-9]", "",salary))

for i in range(22):
    average_salaries[i] = sum_salaries[i] / age_distribution[i]
```

```python
plt.figure(figsize=(30,30))
plt.bar(x,age_distribution)
```

```python
plt.plot(x,sum_salaries)
```

```python
plt.plot(x,average_salaries)
```

```python
np.corrcoef(x,average_salaries)
```
